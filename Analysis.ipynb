```python
import pandas as pd
import os
import ast
import numpy
```

# Gather data into one dataframe

```python
all_results = {}
```

```python
for f in os.listdir('.'):
    if '.data' in f:
        with open(f, 'r') as g:
            results = ast.literal_eval(g.read())
            for key in results:
                if key not in all_results:
                    all_results[key] = [results[key]]
                else:
                    all_results[key].append(results[key])
        # print(f)
```

```python
bacteria_mic_columns = []
for sequence in all_results:
    for result in all_results[sequence]:
        if 'bacteria' in result:
            for bacterium in result['bacteria']:
                bacteria_mic_columns.append(str(bacterium) + '_unit')
                bacteria_mic_columns.append(str(bacterium) + '_value')
bacteria_mic_columns = list(set(bacteria_mic_columns))
```

```python
columns = ['sequence', 'url_source', 'modifications', 'hemolysis', 'references'] + bacteria_mic_columns
```

```python
#df = pd.DataFrame(columns=columns, index=['sequence', 'url_source'])
df = pd.DataFrame(columns=columns,)
```

```python
i = 0
for sequence in all_results:
    for result in all_results[sequence]:
        df_ready = {
            'sequence': sequence,
            'url_source': result['url_sources'][0],
            'modifications': result['modifications'] if 'modifications' in result else None,
            'hemolysis': result['hemolysis'] if 'hemolysis' in result else None,
            'references': result['references'] if 'references' in result else None,
        }
        if 'bacteria' in result:
            for bacteria_and_strain in result['bacteria']:
                df_ready[str(bacteria_and_strain) + '_unit'] = result['bacteria'][bacteria_and_strain]['unit']
                df_ready[str(bacteria_and_strain) + '_value'] = result['bacteria'][bacteria_and_strain]['value']
        df = df.append(df_ready, ignore_index=True)
    print(sequence, i)
    i += 1
```

```python
file_name = 'bacteriadf'
df.to_csv(file_name, sep='\t', encoding='utf-8')
```
